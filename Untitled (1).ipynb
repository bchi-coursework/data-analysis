```python
cancer=[]
chronic_d=[]
hiv=[]
infection_d=[]
injury=[]
life=[]
# finaldata1[(finaldata1[dem] == 1) & (finaldata1['year'] == 2010)]
for c in cat:
    for y in yr:
        if(c=='Cancer'):
            abc=finaldata1[(finaldata1[c] == 1) & (finaldata1['year'] == y)]
            cancer.append(abc.year.count())
        elif c=='chronic_disease' :
            abc=finaldata1[(finaldata1[c] == 1) & (finaldata1['year'] == y)]
            chronic_d.append(abc.year.count())
        elif c=='HIV/AIDS' :
            abc=finaldata1[(finaldata1[c] == 1) & (finaldata1['year'] == y)]
            hiv.append(abc.year.count())
        elif c=='infectious_disease' :
            abc=finaldata1[(finaldata1[c] == 1) & (finaldata1['year'] == y)]
            infection_d.append(abc.year.count())
        elif c=='Injury/Violence' :
            abc=finaldata1[(finaldata1[c] == 1) & (finaldata1['year'] == y)]
            injury.append(abc.year.count())
        elif c=='life_expectancy_and_death_rate_overall' :
            abc=finaldata1[(finaldata1[c] == 1) & (finaldata1['year'] == y)]
            life.append(abc.year.count())
```
